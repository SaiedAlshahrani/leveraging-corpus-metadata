### Token Length Distribution:

```python
aryTokens = arywiki_ordered['total_words'].tolist()

aryDocs = list(range(0, arywiki_ordered.shape[0]))

len(aryTokens), type(aryTokens)
```

```python
aryTokens_ordered = aryTokens.copy()
aryTokens_ordered.sort(reverse=True)
aryTokens_ordered[:5]
```

```python
arzTokens = arzwiki_ordered['total_words'].tolist()

arzDocs = list(range(0, arzwiki_ordered.shape[0]))
len(arzTokens), type(arzTokens)
```

```python
arzTokens_ordered = arzTokens.copy()
arzTokens_ordered.sort(reverse=True)
arzTokens_ordered[:5]
```

```python
arTokens = arwiki_ordered['total_words'].tolist()

arDocs = list(range(0, arwiki_ordered.shape[0]))
len(arTokens), type(arTokens)
```

```python
arTokens_ordered = arTokens.copy()
arTokens_ordered.sort(reverse=True)
arTokens_ordered[:5]
```

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt

figure(figsize=(16, 12), dpi=300)

plt.subplot(3, 3, 1)
plt.title("Moroccan Arabic Wikipedia")
plt.ylabel("Tokens")
plt.xlabel("Tokens Per Article")
plt.plot(aryDocs, aryTokens, label="Length")
plt.axhline(y=arywiki_ordered['total_words'].mean(), color = 'red', linestyle = '-', label="Mean")  
plt.legend(loc='upper left')

plt.subplot(3, 3, 2)
plt.title("Egyptian Arabic Wikipedia")
plt.xlabel("Tokens Per Article")
plt.plot(arzDocs, arzTokens, label="Length")
plt.axhline(y=arzwiki_ordered['total_words'].mean(), color = 'red', linestyle = '-', label="Mean")  
plt.legend(loc='upper left')

plt.subplot(3, 3, 3)
plt.title("Arabic Wikipedia")
plt.xlabel("Tokens Per Article")
plt.plot(arDocs, arTokens, label="Length")
plt.axhline(y=arwiki_ordered['total_words'].mean(), color = 'red', linestyle = '-', label="Mean")  
plt.legend(loc='upper left')

plt.savefig('charts/Token-Length-Distribution.png', bbox_inches='tight')
plt.show()
```
